```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,2*np.pi,0.1)
y = np.sin(x)

plt.xlim(0,2*np.pi)
plt.ylim(-1,1)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
x = np.arange(0,5,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.savefig('sinx.png',bbox_inches="tight", dpi=600)
```
